```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Get the history CSV
training_dir_list = os.listdir('./training/')
training_dir_list.sort()
load_file = './training/' + training_dir_list[-2] + '/history.csv'
history = pd.read_csv(load_file)

# Extract all the historical values
acc = history['acc']
val_acc = history['val_acc']
loss = history['loss']
val_loss = history['val_loss']

# Plot Accuracy
plt.figure()
plt.plot(acc, label='Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.title('Accuracy Over Time - Class Weighting')
plt.xlabel('Epoch #')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig('./visualizations/weight_acc.png')
plt.show()
```

```python
# Plot Loss
plt.figure()
plt.plot(loss, label='Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('Categorical Cross-Entropy Loss Over Time\nClass Weighting')
plt.xlabel('Epoch #')
plt.ylabel('Categorical Cross-Entropy')
plt.legend()
plt.savefig('./visualizations/weight_loss.png')
plt.show()
```
